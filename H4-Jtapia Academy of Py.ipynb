```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Read School and Student Data File and store into Pandas Data Frames
schoold=pd.read_csv("schools_complete.csv",encoding="ISO-8859-1")
studentd=pd.read_csv("students_complete.csv",encoding="ISO-8859-1")
```

```python
schoold.head()
```

```python
studentd.head()
```

```python
#Merge data
datosc=pd.merge(schoold,studentd,how="left", on=["school_name", "school_name"])
```

```python
#DataFcomplete
datosc.head()
```

# District Summary

```python
#Calculate the total number of schools
Escuelas_total=len(datosc["school_name"].unique())
Escuelas_total
```

```python
#Calculate the total number of students
Estudiantes_total=len(datosc["student_name"].unique())
Estudiantes_total
```

```python
#Calculate the total budget
Total_Budget=datosc.groupby("school_name")["budget"].unique().sum()
Total_Budget
```

```python
#Calculate the total budget
Total_Budget.sum()
```

```python
#Calculate the average math score
Math_avg=datosc["math_score"].mean()
Math_avg
```

```python
#Calculate the average reading score
Read_avg=datosc["reading_score"].mean()
Read_avg
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ovavg_score=(Read_avg+Math_avg)/2
ovavg_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
Pstudentsmath=(len(datosc.loc[datosc["math_score"]>=70])/len(datosc))*100
Pstudentsmath
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
Pstudentread=(len(datosc.loc[datosc["reading_score"]>=70])/len(datosc))*100
Pstudentread
```

```python
#Create a dataframe to hold the above results
results={"Total Schools":Escuelas_total,
        "Total Students":Estudiantes_total,
         "Total Budget":[Total_Budget],
         "Average Math Score":[Math_avg],
         "Average Reading Score":[Read_avg],
         "% Passing Math":[Pstudentsmath],
         "% Passing Reading":[Pstudentread],
         "% Overall Passing Rate":[ovavg_score]}
```

# DISTRICT DATA FRAME RESULTS

```python
dfr= pd.DataFrame(results)
dfr
```

# School Summary

```python
Escuelas=datosc.groupby("school_name")["school_name"].unique().count()
Escuelas
```

```python
Type=datosc.groupby("school_name")["type"].unique()
Type
```

```python
#School Name,School Type,Total School Budget
Schoolbud=datosc.groupby(["school_name"])["budget"].unique()
Schoolbud
```

```python
#Total Students
Students=datosc.groupby(["school_name"])["size"].unique()
Students
```

```python
PSB2=Schoolbud/Students
PSB2
```

```python
#Average Math Score
AMS=datosc.groupby(["school_name"])["math_score"].sum()/datosc.groupby(["school_name"])["size"].unique()
AMS
```

```python
#Average Reading Score
ARS=datosc.groupby(["school_name"])["reading_score"].sum()/datosc.groupby(["school_name"])["size"].unique()
ARS
```

```python
#% Passing MATH
passmath=datosc.loc[datosc["math_score"]>=70]
passmath
```

```python
passmath2=passmath["school_name"].value_counts()
#passmath2=passmath["school_name"].value_counts()/passmath["size"].unique()
#passmath2
passmath2
```

```python
passmat3=passmath.groupby("school_name")["size"].unique()
passmat3
```

```python
passmatvf=(passmath2/passmat3*100)
passmatvf
```

```python
#% Passing READ
passread=datosc.loc[datosc["reading_score"]>=70]
passread
```

```python
passreadf=passread["school_name"].value_counts()/passread.groupby("school_name")["size"].unique()*100
passreadf
```

```python
Overall=(passmatvf+passreadf)/2
Overall
```

# School Summary

```python
sumschool={
    'School Type': Type,
    'Total Students':Students,
    'Total School Budget': Schoolbud,
    'Per Student Budget': PSB2,
    'Average Math Score': AMS,
    'Average Reading Score': ARS,
    '% Passing Math': passmatvf,
    '% Passing Reading': passreadf,
    '% Overall Passing Rate': Overall,}
```
