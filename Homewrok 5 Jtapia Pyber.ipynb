```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the City and Ride Data
cityd=pd.read_csv("data_city_data.csv")
rided=pd.read_csv("data_ride_data.csv")
```

```python
cityd.head()
```

```python
rided.head()
```

```python
mfare=rided.groupby("city")["fare"].sum()
mfare
```

```python
trip=rided.groupby("city")["ride_id"].count()
trip
```

```python
dfc=pd.merge(cityd,rided,how="outer", on=["city"])
```

```python
dfc.count()
```

```python
dfc.head()
```

```python
dfc.dtypes
```

```python
types=dfc.groupby("city")["type"].unique()
types
```

```python
mnfare=dfc.groupby(["city"])["fare"].sum()/dfc.groupby(["city"])["ride_id"].count()
mnfare
```

```python
farec=dfc.groupby(["city"])["ride_id"].count()
farec
```

```python
ndriver=dfc.groupby("city")["driver_count"].unique()
ndriver
```

```python
nride=dfc.groupby("city")["ride_id"].count()
nride
```

```python
pyber={'Type of City':types,
      'Fare mean':mnfare,
      'Number of drivers':ndriver,
      'Number of rides':nride,
      'Number of fares':farec}
```

```python
dfpyber=pd.DataFrame(pyber)
dfpyber
```

```python
dfpyber.dtypes
```

```python
dfpyber["Number of drivers"]=dfpyber["Number of drivers"].astype(np.int64)
dfpyber["Type of City"]=dfpyber["Type of City"].astype(str)
```

```python
dfpyber.dtypes
```

```python
dfpyber["Type of City"].value_counts()
```

```python
dfpyber["Type of City"]=dfpyber["Type of City"].replace({"['Urban']":"Urban","['Suburban']":"Suburban","['Rural']":"Rural"})
```

```python
dfpyber["Type of City"].value_counts()
```

```python
dfpyber["Drivers2"]=dfpyber["Number of drivers"]*dfpyber["Number of drivers"]
```

# Bubble Plot of Ride Sharing Data

```python
dfpyber
```

```python
dfpyber.dtypes
```

```python
x=dfpyber["Number of rides"][dfpyber["Type of City"]=='Suburban'],
y=dfpyber["Fare mean"][dfpyber["Type of City"]=='Suburban'],
#z=dfpyber["Number of drivers"][dfpyber["Type of City"]=='Suburban'],
z=dfpyber["Drivers2"][dfpyber["Type of City"]=='Suburban'].count(),
plt.scatter(x, y, s=z, c=x, cmap="Wistia", alpha=1, edgecolors="gold", linewidth=2)
plt.xlabel("Total of Rides per city,Suburban Cities")
plt.ylabel("Average fare ($)")
plt.title("Ride Sharing on Suburban Cities")
plt.show()
```

```python
x=dfpyber["Number of rides"][dfpyber["Type of City"]=="Urban"],
y=dfpyber["Fare mean"][dfpyber["Type of City"]=="Urban"],
z=dfpyber["Drivers2"][dfpyber["Type of City"]=="Urban"].count(),
plt.scatter(x, y, s=z, c=x, cmap="Blues", alpha=0.6, edgecolors="skyblue", linewidth=2)
plt.xlabel("Total of Rides per city, Urban Cities")
plt.ylabel("Average fare ($)")
plt.title("Ride Sharing on Urban Cities")
plt.show()
```

```python
x=dfpyber["Number of rides"][dfpyber["Type of City"]=="Rural"],
y=dfpyber["Fare mean"][dfpyber["Type of City"]=="Rural"],
z=dfpyber["Drivers2"][dfpyber["Type of City"]=="Rural"].count(),
plt.scatter(x, y, s=z, c=x, cmap="Oranges", alpha=2, edgecolors="coral", linewidth=2)
plt.xlabel("Total of Rides per city, Rural Cities")
plt.ylabel("Average fare ($)")
plt.title("Ride Sharing on Rural Cities")
plt.show()
```

```python

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np
#type=dfpyber.groupby("Type of City")
colors = ("gold", "coral", "skyblue")
def attribute_color(typ):
    #colors ={(dfpyber["Type of City"]=="Rural":"gold"),(dfpyber["Type of City"]=="Suburban":"coral"),(dfpyber["Type of City"]=="Urban":"skyblue")}
    colors={"Rural":"gold","Suburban":"skyblue","Urban":"coral"}
    return colors.get(typ,'white')
color_type=list()
qty_type=len(dfpyber["Type of City"])
for i in range(qty_type):
    color_type.append(attribute_color(dfpyber["Type of City"][i]))
color_type
                      

    
```

```python
 dfpyber.groupby("Type of City")["Number of rides"].head()

```

```python
plt.scatter(x = dfpyber["Number of rides"],
            y = dfpyber["Fare mean"],
            s = dfpyber["Drivers2"],
            c=color_type,
            alpha = 0.6,
           edgecolors="white")
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average fare ($)")
plt.figtext(.95, .6,"Note:Cicle size correlate with driver count per city")
plt.grid(b=True,which='major',color="gray",axis='both',)
city=["Rural","Urban","Suburban"]
legend = list()
for x in range(len(colors)):
    legend.append(mlines.Line2D([0], [0],
                                        linestyle='none',
                                        marker='o',
                                        alpha=0.6,
                                        markersize=6,
markerfacecolor=colors[x]))
legend1 = plt.legend(legend,
                     city,
                     numpoints=1,
                     fontsize=10,
                     loc='upper right',
                     shadow=True)
```

# Total RIDES City Type

```python
Rbycity=dfpyber.groupby("Type of City")["Number of rides"].sum()/dfpyber["Number of rides"].sum()*100
Rbycity
```

```python
categ = ["Rural", "Suburban", "Urban"]
slices = Rbycity
colors = ["coral","skyblue","gold",]
explode = (.05, .05, .05)
```

```python
plt.pie(slices,explode=explode,labels=categ,colors=colors,autopct='%1.1f%%',shadow=True,startangle=180,textprops={'color':"w",'fontsize': 14},)
plt.axis("equal")
plt.title("Percentage of Rides by type of city")
```
